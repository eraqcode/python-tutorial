# **Calc Module**

```python
def get_add(a, b):

    """Addition Function"""

    return a + b
```

```python
def get_substract(a, b):

    """Substract Function"""

    return a - b
```

```python
def get_multiply(a,b):

    """Multiply Function"""

    return a * b
```

```python
def get_divide(a, b):

    """Divide Function"""

    if b == 0 :

        raise ZeroDivisionError("Can not divide by zero")

    return a / b
```
